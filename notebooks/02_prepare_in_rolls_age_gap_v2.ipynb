```python
import pandas as pd
from glob import glob
import os
import time
```

```python
CHUNK_SIZE = 1000000
```

```python
sex_remap = {'महिला': 'Female', 'पुरूष': 'Male', 'स्त्री': 'Female'}
has_husband = {'father': 'no', 'husband': 'yes'}
```

```python
%%time
for fn in sorted(glob('/opt/data2/parse-in-rolls/clean+t13n/*.gz')):
    print(fn)
    output_file = '/opt/data2/in_rolls/age_gap/' + os.path.basename(fn).split('_')[0] + '.csv.gz'
    #output_file = '/opt/data2/in_rolls/age_gap/' + os.path.basename(fn).split('_')[0] + '.csv'
    df = pd.read_csv(fn, nrows=1)
    cols = ['age', 'sex', 'state', 'year', 'relationship', 'id', 'house_no']
    if 'elector_name_t13n' in df.columns:
        cols.extend(['elector_name_t13n', 'father_or_husband_name_t13n', 'polling_station_name_t13n'])
    else:
        cols.extend(['elector_name', 'father_or_husband_name', 'polling_station_name'])
    count = 0
    total = 0
    header = True    
    for adf in pd.read_csv(fn, usecols=cols, chunksize=CHUNK_SIZE):
        t1 = time.time()
        adf.state = adf.state.str.lower()
        adf.rename(columns={'elector_name_t13n': 'elector_name',
                            'father_or_husband_name_t13n': 'father_or_husband_name',
                            'polling_station_name_t13n': 'polling_station_name',
                            'relationship': 'has_husband'}, inplace=True)
        # V2: finding-couples script don't like age in float format
        adf.age = adf.age.fillna(0).astype(int)
        adf.sex.replace(sex_remap, inplace=True)
        #print(adf.sex.unique())
        adf['has_husband'] = adf.has_husband.str.lower()
        adf.has_husband.replace(has_husband, inplace=True)
        adf['change'] = None
        # V2: Clean non-alpha from name
        adf['elector_name'] = adf.elector_name.str.replace('[^a-zA-Z\s]', '', regex=True).str.strip()
        adf['father_or_husband_name'] = adf.father_or_husband_name.str.replace('[^a-zA-Z\s]', '', regex=True).str.strip()
        adf = adf[['id', 'elector_name', 'father_or_husband_name', 'has_husband', 'house_no', 'age', 'sex', 'year', 'polling_station_name', 'change', 'state']]
        #print(fn, adf.columns)
        adf.to_csv(output_file, mode='a', header=header, index=False, compression='gzip')
        #adf.to_csv(output_file, mode='a', header=header, index=False)
        header = False
        count += 1
        total += len(adf)
        t2 = time.time()
        print(count, total, t2 - t1, fn)
        #break
    #break
```
